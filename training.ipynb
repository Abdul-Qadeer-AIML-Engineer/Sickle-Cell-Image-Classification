# Evaluation

```python
# testing_negatives = "Final_Data/test/Negative"
# testing_positives = "Final_Data/test/Positive"
# model_path = "model1.pt"

import os
import torch
from ultralytics import YOLO
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import numpy as np

# Paths
testing_negatives = "Final_Data/test/Negative"
testing_positives = "Final_Data/test/Positive"
model_path = "model1.pt"

def test_model(model_path, neg_dir, pos_dir):
    # Load the trained model
    model = YOLO(model_path)
    
    # Check if GPU is available
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.model.to(device)
    print(f"Using device: {device}")

    # Lists to store true labels and predictions
    true_labels = []
    pred_labels = []

    # Process Negative class (label 0)
    neg_files = [os.path.join(neg_dir, f) for f in os.listdir(neg_dir) if f.endswith(('.jpg', '.jpeg', '.png'))]
    for img_path in neg_files:
        # Predict
        results = model.predict(img_path, imgsz=640)
        pred_class = results[0].probs.top1  # Get the predicted class index (0 or 1)
        
        # Store results
        true_labels.append(0)  # True label for Negative
        pred_labels.append(pred_class)

    # Process Positive class (label 1)
    pos_files = [os.path.join(pos_dir, f) for f in os.listdir(pos_dir) if f.endswith(('.jpg', '.jpeg', '.png'))]
    for img_path in pos_files:
        # Predict
        results = model.predict(img_path, imgsz=640)
        pred_class = results[0].probs.top1  # Get the predicted class index (0 or 1)
        
        # Store results
        true_labels.append(1)  # True label for Positive
        pred_labels.append(pred_class)

    # Calculate metrics
    accuracy = accuracy_score(true_labels, pred_labels)
    precision = precision_score(true_labels, pred_labels)
    recall = recall_score(true_labels, pred_labels)
    f1 = f1_score(true_labels, pred_labels)
    conf_matrix = confusion_matrix(true_labels, pred_labels)

    # Print results
    print("\n=== Test Results ===")
    print(f"Total Images Tested: {len(true_labels)} (Negative: {len(neg_files)}, Positive: {len(pos_files)})")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print("\nConfusion Matrix:")
    print(conf_matrix)
    print("[[TN  FP]")
    print(" [FN  TP]]")

    return true_labels, pred_labels, conf_matrix

test_model(
    model_path=model_path,
    neg_dir=testing_negatives,
    pos_dir=testing_positives
)
```

# Plotting

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def display_single(image_path, title=None):
    img = mpimg.imread(image_path)
    plt.figure(figsize=(10, 8))  # Set figure size (width, height)
    plt.imshow(img)

    if title:
        plt.title(title, fontsize=12)

    plt.axis('off')
    plt.show()

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def display_double(img_path1, img_path2, title1="Image 1", title2="Image 2"):
    img1 = mpimg.imread(img_path1)
    img2 = mpimg.imread(img_path2)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))  # Adjust figsize for overall width/height
    ax1.imshow(img1)
    ax1.set_title(title1, fontsize=12, pad=10)
    ax1.axis('off')  # Hide axes
    
    ax2.imshow(img2)
    ax2.set_title(title2, fontsize=12, pad=10)
    ax2.axis('off')  # Hide axes
    plt.tight_layout(pad=2.0)
    plt.show()
```
